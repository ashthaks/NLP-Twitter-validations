```python
df.drop(columns=['id','location'],inplace=True)
df
```

```python
df=df.loc[df['target']!='Irrelevant']
df
```

```python
df.reset_index(inplace=True,drop=True)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
# p= 1
# n= 0
# neg= -1

df['target']=df['target'].map({'Positive':1,'Neutral':0,'Negative':-1})
df
```

```python
tweets=df.text
tweets
```

```python
tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')
tweets
```

```python
import nltk
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
```

```python
tweets=tweets.apply(lambda x:[stemmer.stem(token.lower()) for token in tk.tokenize(x)]).apply(lambda token:" ".join(token))
tweets
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
tweets=tweets.apply(lambda x:[token for token in tk.tokenize(x) if token not in sw]).apply(lambda token:" ".join(token))
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
X=vec.fit_transform(tweets)
X
```

```python
print(X)
```

```python
vec.get_feature_names_out()
```

```python
X.shape
```

```python
y=df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
sv=SVC()
knn=KNeighborsClassifier()
ad=AdaBoostClassifier(random_state=1)
rf=RandomForestClassifier(random_state=1)
lst=[sv,knn,ad,rf]
for li in lst:
  print('****************',li,'**********************')
  li.fit(X_train,y_train)
  y_pred=li.predict(X_test)
  print(classification_report(y_test,y_pred))
```
